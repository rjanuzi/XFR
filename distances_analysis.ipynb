```python
import seaborn as sns
import pandas as pd
import json
from dataset import get_file_path, DATASET_KIND_MORPH, DATASET_KIND_ALIGNED
import matplotlib.pyplot as plt
from matplotlib import rcParams
from mpl_toolkits.axes_grid1 import ImageGrid
from pathlib import Path
```

```python
distances = json.load(open("fr\\distances.json", "r"))
distances_lst = []
for pair, tmp_distances in distances.items():
    tmp_distances['pair'] = pair
    distances_lst.append(tmp_distances)

df = pd.DataFrame(distances_lst)
df.loc[:, "pair"] = df.loc[:, 'pair'].apply(lambda x: x.split(' x '))
df.loc[:, "img1"] = df.loc[:, 'pair'].apply(lambda x: x[0])
df.loc[:, "img2"] = df.loc[:, 'pair'].apply(lambda x: x[1])
```

```python
# plt.matshow(df.corr())
f = plt.figure(figsize=(19, 15))
plt.matshow(df.corr(), fignum=f.number)
plt.xticks(range(df.select_dtypes(['number']).shape[1]), df.select_dtypes(['number']).columns, fontsize=14, rotation=45)
plt.yticks(range(df.select_dtypes(['number']).shape[1]), df.select_dtypes(['number']).columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16)
```
